```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
# Load MNIST dataset and concatenate training and testing data
(X_train_full, y_train_full), (X_test_full, y_test_full) = mnist.load_data()

# Combine the original training and test datasets
X_full = np.concatenate((X_train_full, X_test_full))
y_full = np.concatenate((y_train_full, y_test_full))

# Train-test split with a custom split ratio (70% for training)
train_size = 0.7
X_train, X_test, y_train, y_test = train_test_split(X_full, y_full, train_size=train_size, random_state=2019)

# Normalize the data
X_train = X_train.reshape(-1, 28*28).astype('float32') / 255.0
X_test = X_test.reshape(-1, 28*28).astype('float32') / 255.0

# Convert labels to categorical one-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
def create_model(hidden_layers=2, neurons_per_layer=64, learning_rate=0.001):
    model = models.Sequential()
    model.add(layers.InputLayer(input_shape=(28*28,)))  # Input layer

    # Add hidden layers
    for _ in range(hidden_layers):
        model.add(layers.Dense(neurons_per_layer, activation='relu'))

    model.add(layers.Dense(10, activation='softmax'))  # Output layer (10 classes)

    # Compile the model
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
def plot_learning_curves(history, title="Learning Curves"):
    # Plot training & validation accuracy values
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(title)
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()

    # Plot training & validation loss values
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Loss Curves')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()

# Train and evaluate the model with different hyperparameters
hidden_layers = 2
neurons_per_layer = 128
learning_rate = 0.001

model = create_model(hidden_layers=hidden_layers, neurons_per_layer=neurons_per_layer, learning_rate=learning_rate)

history = model.fit(X_train, y_train, epochs=20, batch_size=128, validation_data=(X_test, y_test))

# Plot learning curves
plot_learning_curves(history, title=f"Learning Curves (LR={learning_rate}, Hidden Layers={hidden_layers}, Neurons={neurons_per_layer})")
```

```python
# Train the model with a higher learning rate
model_high_lr = create_model(hidden_layers=2, neurons_per_layer=128, learning_rate=0.01)
history_high_lr = model_high_lr.fit(X_train, y_train, epochs=20, batch_size=128, validation_data=(X_test, y_test))

# Plot learning curves for high learning rate
plot_learning_curves(history_high_lr, title="Learning Curves (High Learning Rate)")
```

```python
# Train the model with more hidden layers
model_more_layers = create_model(hidden_layers=4, neurons_per_layer=128, learning_rate=0.001)
history_more_layers = model_more_layers.fit(X_train, y_train, epochs=20, batch_size=128, validation_data=(X_test, y_test))

# Plot learning curves for more hidden layers
plot_learning_curves(history_more_layers, title="Learning Curves (More Hidden Layers)")
```

```python


# Apply PCA for dimensionality reduction
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.reshape(-1, 28*28))
X_test_scaled = scaler.transform(X_test.reshape(-1, 28*28))

pca = PCA(n_components=0.95)  # Retain 95% of variance
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

def create_model_with_regularization(hidden_layers=2, neurons_per_layer=64, learning_rate=0.001, l2_reg=0.01):
    model = models.Sequential()
    model.add(layers.InputLayer(input_shape=(X_train_pca.shape[1],)))  # Input layer

    # Add hidden layers with L2 regularization
    for _ in range(hidden_layers):
        model.add(layers.Dense(neurons_per_layer, activation='relu',
                               kernel_regularizer=tf.keras.regularizers.l2(l2_reg)))

    model.add(layers.Dense(10, activation='softmax'))  # Output layer (10 classes)

    # Compile the model
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Train the model with regularization and PCA
model_with_reg = create_model_with_regularization(hidden_layers=2, neurons_per_layer=128,
                                                   learning_rate=0.001, l2_reg=0.01)

history_with_reg = model_with_reg.fit(X_train_pca, y_train, epochs=20, batch_size=128, validation_data=(X_test_pca, y_test))

# Plot learning curves with regularization and PCA
plot_learning_curves(history_with_reg, title="Learning Curves (With Regularization and PCA)")

# Train the model without PCA and regularization for comparison
model_without_reg = create_model(hidden_layers=2, neurons_per_layer=128,
                                  learning_rate=0.001)

history_without_reg = model_without_reg.fit(X_train.reshape(-1, 28*28), y_train,
                                             epochs=20, batch_size=128, validation_data=(X_test.reshape(-1, 28*28), y_test))

# Plot learning curves without regularization and PCA
plot_learning_curves(history_without_reg, title="Learning Curves (Without Regularization and PCA)")
```
